### 02_06 오후수업

```python
#strip() : 문자열 양 옆에 공백을 지워주는 함수
s = "             안녕하세요 여러분!          "
s.strip()
#print(s)
```

```python
# replace('문자1','문자2') : 문자1을 문자2로 바꿔주는 함수
s = s.replace('여러분','3반 친구들').strip()  # (한방에 쓰기 가능)함수를 계속해서 이어나가면 된다.
```

```python
s
```

```python
#split() : 함수에 넣어준 문자를 기준으로 문자열을 나눠준다.
s.split(" ") # 띄어쓰기 기준으로 나뉨
```

```python
s.split(" ")[0] # 쪼개진것에 인덱스 번호가 붙어서 0열을 가져와줌
```

```python
# 1. 초로 된 시간을 입력받기
time = int(input("시간 입력 >> "))
# 2. 시간 구하기
hour = time // 3600
# 3. 분 구하기
minute = time % 3600 // 60
# 4. 초구하기
second = time % 3600 %60

print("{}시간 {}분 {}초".format(hour, minute, second))
```
