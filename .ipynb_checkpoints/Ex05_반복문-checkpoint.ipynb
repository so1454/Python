```python
while True:
    a = int(input("첫번째 정수 입력 >> "))
    b = int(input("두번째 정수 입력 >> "))
    c = a+b
    if a == 0 & b == 0 :
        print("프로그램이 종료되었습니다.")
        break
    else : 
        print(f"두 정수의 합 : {c}")
```

```python
# 현재 몸무게 / 목표 몸무게 입력받기!

nowKg = int(input("현재 몸무게 >> "))
goalKg = int(input("목표 몸무게 >> "))
week = 1
minusKg = 0

while True :
    minusKg = int(input(f"{week}주차 감량 몸무게>> "))
    week += 1
    nowKg -= minusKg
    if nowKg < goalKg :
        print(f"{nowKg}kg 달성!! 축하합니다")
        break
```

```python
import random
```

```python
num = random.randint(1,50)

while True :
    a = int(input("숫자를 입력하세요 >> "))
    if a > num :
        print(f"{a}보다 작은 수 입니다.")
    elif a < num :
        print(f"{a}보다 큰 수 입니다.")
    elif a == num :
        print("정답을 맞추셨습니다.")
        break
```

```python
score_list = [90,45,70,60,55]
student = 0

for jung in score_list :
    student += 1
    if jung >= 60 :
        print(f"{student}번 학생은 합격입니다")
    else :
        print(f"{student}번 학생은 불합격입니다")
```

```python
python_score = [57, 86, 63, 92, 35, 79]
student = 0
result = 0
for python in python_score :
    student += 1
    if python >= 80 :
        result = "A"
    elif python >= 60 :
        result = "B"
    else :
        result = "C"
    print(f"{student}번은 {python}점이며, {result}등급 입니다.")
```

```python
# range(시작숫자, 끝 숫자 +1, 증감량(생략가능)) : 
# 시작값과 증감량은 생략이 가능하다!
# 시작값이 생략된 경우 숫자는 몇부터? -> 0부터 시작한다.
# 증감량이 생략된 경우는 몇씩 숫자가 변할까? -> 1씩 더해진다.
for i in range(1,10,1) :
    print(i)
```

```python
# print()함수의 속성
# end 속성
for i in range(1,10,1) :
    print(i,end=' ')  # --> 기본적으로 \n(개행이 들어가있다)
```

```python
list2 = [4,5,2,1,99,15,2,7,27]
a = 0

# 내가한거
#얘만 다름!
for a in list2 :
    a

for i in list2 :
    if i > a :
        a = i
        print(a)
```

```python
#쌤이 해주신거        
max_num = list2[0]

for i in list2 :
    if i>max_num :
        max_num = i
print(max_num)
```

```python
# 젤 작은 수 구하기
min_num = list2[0]

for i in list2 :
    if i<min_num :
        min_num = i
print(min_num)
```
